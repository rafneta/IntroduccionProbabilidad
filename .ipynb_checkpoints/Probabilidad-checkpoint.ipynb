#  Gráficas en Python


Como es natural de pensar no podemos generar en una computadora conjuntos continuos (pues necesitariamos cantidad infinita de memoria) así que si queremos general gráficas de funciones cuyo dominio es un conjunto continuo, tomamos unas muestras (un subconjunto finito lo suficientemente "bueno") y aplicamos la función, posteriormente con ayuda de la represenación visual emulamos la aplicación de la función a un conjunto infinito. ##   Función de densidad $f(x)$ VA continuas  

- $P(X\in[a,b])=\int_a^b f(x) dx $
- $f(x)\geq 0$
- $\int\limits_{-\infty}^\infty f(x)=1$

Como se puede observar generar una gráfica con un "buen" formato implica editar varias caracteristicas, así que será una buena práctica generar funciones para gráficar de manera automática. A continuación cargamos el módulo donde se ha programado la función en cuestión, el nombre es **gfdvac** (**g**ráfica **f**unción **d**ensidad **v**ariable **a**leatoria **c**ontinua)

```python
#Cargamos el modulo 
from ProgramasProba.Proba import *

#Para el ejemplo anterior

def fejemplo2(x):
    return np.piecewise(x, [x < 0, (0<=x) & (x<3), x>3 ],\
                        [0,lambda x: (2.0/9.0)*x,0])


gfdvac((-2,10),fejemplo2)
```

##  Función de distribución (acumulada) $F(x)$ VA continuas  

- $F(X\leq x)=P(X\in(-\infty,x])=\int\limits_{-\infty}^x f(x) dx $
- $f(x)=\frac{dF(x)}{dx}$

Se ha generado una función, gFdfvac (**g**ráfica **F**(x) **d**istribución de **f** **v**ariable **a**leatoria **c**ontinua ) que esta contenida en el módulo previamente cargado (ProgramasProba), que calcula $F(x)$ de manera númerica (método númerico para integrar $f(x)$ y obtener $F(x)$), el método (función quad) se encuentra en el submódulo *Integrate* del módulo *Scipy* de Python. Se obtiene la grafica de $F(x)$ a partir de conocer $f(x)$.

-[Documento Scipy](https://docs.scipy.org/doc/scipy/reference/tutorial/index.html). Módulo para computación cientifica compatible con Numpy, Scipy, Matplotlib, entre otros 


```python
gFdfvac((-1.5,4),fejemplo2)
```

 ### Funciones de probabilidad  $f(x)$ (función de masa de probabilidad) VA discretas

- $P(X=x)=f(x)$
- $f(x)\geq 0$
- $\sum\limits_{x}f(x)=1$

Se muestra la idea para implementar funciones con estas gráficas 

```python
X=np.array([0,1,2])
f=np.array([0.25,0.5,0.25])

tam_marcador=12
marcador='o'
fig=plt.figure(figsize=(8,4))
ax=plt.axes()

ax.plot(X,f,'bo',ms=tam_marcador)
ax.vlines(X, [0], f,colors='b', lw=5, alpha=0.3)
plt.xticks(X)
plt.yticks(f)

X=np.array([0,1,2])
f=np.array([0.25,0.5,0.25])
```

Se utiliza la función **gfpvad** (**grafica** **f**uncion de **p**robabilidad **v**ariable **a**leatoria **d**iscreta) cuyos datos son los valores de la variable aleatoria **X** en formato de array, y sus correspondientes valores de probabiidad **f**, tambien en formato de array. Esta función se encuetra en el modulo importado. 

```python
X=np.array([0,1,2])
f=np.array([0.25,0.5,0.25])

gfpvad(X,f)
#guardar=True
```

##  Función de distribución (acumulada) $F(x)$ VA discretas  

- $F(X\leq x)=P(X\in(-\infty,x])=\sum\limits_{u\leq x} f(u)$
- $f(x)=F(x)-F(x^{-})$

Se ha generado una función, gFdfvac (**g**ráfica **F**(x) **d**istribución de **f** **v**ariable **a**leatoria **d**iscreta) que esta contenida en el módulo previamente cargado (ProgramasProba). Se obtiene la grafica de $F(x)$ a partir de conocer  $X$ y $f(x)$ en sus formatos de array.

```python
X=np.array([0,1,2])
f=np.array([0.25,0.5,0.25])

gFdfvad(X,f)
```

# Variables aleatorias discretas
[Notación de scipy.stats](https://docs.scipy.org/doc/scipy/reference/tutorial/stats/discrete.html)

- *pmf(x)* (Probability Mass Function) Función de densidad $f(x)=F(x)-F(x^{-})$ 
- *cdf(x)* (Cumulative Distribution Function) Funcion de distribución $F(x)$
- *ppf(x)* (Percent Point Function) Función inversa a cdf(x) $G(q)=F^{-1}(q)$


[Distribuciones discretas](https://docs.scipy.org/doc/scipy/reference/tutorial/stats/discrete.html#discrete-distributions-in-scipy-stats)

## Distribución uniforme discreta

```python
X=np.array(range(-1,4))
vduniforme=stats.randint(low=-1,high=4)
f=vduniforme.pmf(X)
gfpvad(X,f)
gFdfvad(X,f)
print(vduniforme.mean())
print(vduniforme.median())
print(vduniforme.var())
print(vduniforme.std())
```

## Distribución Bernoulli  (es discreta)

```python
X=np.array([0,1])
vdbernoulli=stats.bernoulli(p=0.1)
f=vdbernoulli.pmf(X)
gfpvad(X,f)
gFdfvad(X,f)
print(vdbernoulli.mean())
print(vdbernoulli.median())
print(vdbernoulli.var())
print(vdbernoulli.std())
```

## Distribución binomial  (es discreta)

```python
X=np.array(range(0,10))
vdbinomial=stats.binom(n=10,p=0.5)
f=vdbinomial.pmf(X)
gfpvad(X,f,etiquetas=False)
gFdfvad(X,f,etiquetas=False)
print(vdbinomial.mean())
print(vdbinomial.median())
print(vdbinomial.var())
print(vdbinomial.std())
```

```python
#EJEMPLO 1
vdbinomial=stats.binom(n=4,p=0.75)
vdbinomial.pmf(2)
print("ejemplo 1: ",vdbinomial.pmf(2),27/128.0)

#EJEMPLO 2
vdbinomial=stats.binom(n=15,p=0.4)
#a)
X=np.array(range(10,16))
print("ejemplo 2a): ",np.sum(vdbinomial.pmf(X)),1-vdbinomial.cdf(9))
#b)
X=np.array(range(3,9))
print("ejemplo 2b): ",np.sum(vdbinomial.pmf(X)),vdbinomial.cdf(8)-vdbinomial.cdf(2))
#c)
print("ejemplo 2c): ",vdbinomial.pmf(5),vdbinomial.cdf(5)-vdbinomial.cdf(4))

#EJEMPLO 3
#a)
vdbinomial=stats.binom(n=20,p=0.03)
print("ejemplo 3a): ",1-vdbinomial.pmf(0))
#b)
vdbinomial=stats.binom(n=10,p=0.4562)
print("ejemplo 3b): ",vdbinomial.pmf(3))

#EJEMPLO 4
#a)
vdbinomial=stats.binom(n=10,p=1.0/3)
X=np.array(range(6,11))
print("ejemplo 4): ",np.sum(vdbinomial.pmf(X)),1-vdbinomial.cdf(5))
```

## Distribución geométrica  (es discreta)

```python
X=np.array(range(0,10))
#la f(x) esta desplazada a la derecha en la definción en python respecto a nuestra definición. 
vdgeometrica=stats.geom(p=0.4,loc=-1)# para que coincida con nuestra definición loc=-1
f=vdgeometrica.pmf(X)
gfpvad(X,f,etiquetas=False)
gFdfvad(X,f,etiquetas=False)
print(vdgeometrica.mean())
print(vdgeometrica.median())
print(vdgeometrica.var())
print(vdgeometrica.std())
```

# Variables aleatorias continuas

[Notación de scipy.stats](https://docs.scipy.org/doc/scipy/reference/tutorial/stats/continuous.html)

- *pdf(x)* (Probability Density Function) Función de densidad $f(x)=F'(x)$ 
- *cdf(x)* (Cumulative Distribution Function) Funcion de distribución $F(x)$
- *ppf(x)* (Percent Point Function) Función inversa a cdf(x) $G(q)=F^{-1}(q)$


[Distribuciones continuas](https://docs.scipy.org/doc/scipy/reference/tutorial/stats/continuous.html#continuous-distributions-in-scipy-stats)



## Distribución Uniforme continua

```python
# Ya se ha importando scipy.stats
#a=loc, b=loc+scale
vcuniforme = stats.uniform(loc=0, scale=2)
f = vcuniforme.pdf
F = vcuniforme.cdf
gfdvac((-1,3), f) #de acuerdo a como se programo 
gFdfvac((-1,3), f)
gFdvac((-1,3),F)
print(vcuniforme.mean())
print(vcuniforme.median())
print(vcuniforme.var())
print(vcuniforme.std())
```

## Distribución Exponencial

```python
# Ya se ha importando scipy.stats
#loc=0, scale=1/lambda
#para lambda=1/2
vcexponencial = stats.expon(scale=2)
f = vcexponencial.pdf
F = vcexponencial.cdf
gfdvac((-2,8),f) #de acuerdo a como se programo 
gFdvac((-5,10),F)
#gFdfvac((-1,3), f)
print(vcexponencial.mean())
print(vcexponencial.median())
print(vcexponencial.var())
print(vcexponencial.std())
```

## Distribución Gama (es continua)

```python
#loc=0, a=\alpha, scale=1/lambda
#Para lambda=1/2
vcgama = stats.gamma(a=3, loc=0, scale=2)
f = vcgama.pdf
F = vcgama.cdf
gfdvac((-1,20),f) #de acuerdo a como se programo 
gfdvac((-1,20),F)
#gFdfvac((-1,20),f)
print(vcgama.mean())
print(vcgama.median())
print(vcgama.var())
print(vcgama.std())
```

## Distribución Weibull (es continua)

```python
#loc=0, c=alpha, scale=1/lambda
vcWeibull = stats.weibull_min(c=2, loc=0, scale=1)
f = vcWeibull.pdf
F = vcWeibull.cdf
gfdvac((0,7),f) #de acuerdo a como se programo 
gFdvac((0,7),F)
#gFdfvac((-1,10),f)
print(vcWeibull.mean())
print(vcWeibull.median())
print(vcWeibull.var())
print(vcWeibull.std())
```

## Distribución Normal (es continua)


```python
# Ya se ha importando scipy.stats
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=3, scale=2)
f = vcnormal.pdf
F = vcnormal.cdf
gfdvac((-2,8),f) #de acuerdo a como se programo 
gFdvac((-2,8),F)
#gFdfvac((-5,10),f)
print(vcnormal.mean())
print(vcnormal.median())
print(vcnormal.var())
print(vcnormal.std())
```

```python
#EJEMPLO 1
# a)
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=0, scale=1)
F = vcnormal.cdf
print("ejemplo 1a): ",1-F(1.84),F(-1.84))
# b)
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=0, scale=1)
F = vcnormal.cdf
print("ejemplo 1b): ",F(0.86)-F(-1.97))

#EJEMPLO 2
# a)
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=0, scale=1)
Finv = vcnormal.ppf
print("ejemplo 2a): ",Finv(0.6985))
# b)
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=0, scale=1)
Finv = vcnormal.ppf
print("ejemplo 2b): ",Finv(F(-0.18)-0.4197))


#EJEMPLO 3
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=50, scale=10)
F = vcnormal.cdf
vcnormal2 = stats.norm(loc=0, scale=1)
F2 = vcnormal2.cdf
print("ejemplo 3): ",F(62)-F(45),F2(1.2)-F2(-0.5))


#EJEMPLO 4
#a)
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=0, scale=1)
Finv = vcnormal.ppf
print("ejemplo 4a): ",Finv(0.45)*6+40)
#b)
# loc=mu, scale=sigma
vcnormal = stats.norm(loc=0, scale=1)
Finv = vcnormal.ppf
print("ejemplo 4b): ",Finv(0.86)*6+40)
```
