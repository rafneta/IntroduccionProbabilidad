## Teoría de conjuntos.

Un **conjunto** $A$ es una colección de objetos que son los **elementos** de $A$. Escribimos $\omega\in A$ para denotar que $\omega$ **pertenece** a $A$ o que $\omega$ es elemento de $A$. Escribimos $\omega \notin A$ en caso contrario. Algunas consideraciones 






<div>
Definicion espacio muetral    
</div>

* Los conjuntos los denotamos con $\lbrace \rbrace $

* dentro de estas _llaves_ escribimos los elementos separados por comas $\lbrace 1,2 \rbrace$

* cada elemento debe aparecer solo una vez, si aparece más de una vez se omiten las repeticiones $\lbrace 1,1,1,2 \rbrace=\lbrace 1,2 \rbrace$

* el orden en que aparecen los elemntos no es importante $\lbrace 1,2 \rbrace = \lbrace 2,1 \rbrace$

**Ejemplos:**

- $\mathbb{N}=\lbrace 1, 2 ,3, \ldots\rbrace$ el conjunto de los numeros naturales ($\mathbb{N}=\lbrace 0, 1, 2 ,3, \ldots\rbrace$ depende del autor). $2\in \mathbb{N}$, $126\in \mathbb{N}$, $-5\notin \mathbb{N}$, $\frac{3}{4}\notin \mathbb{N}$. $\sqrt{2}\notin \mathbb{N}$, $\pi\notin \mathbb{N}$

- $\mathbb{Z}=\lbrace \ldots,-2,-1,0,1, 2 ,3, \ldots\rbrace$

- $\mathbb{Q}=\lbrace \frac{a}{b}\mid a,b\in\mathbb{Z}, b\neq 0, \rbrace$

- Los números reales $\mathbb{R}$

Observemos que los conjuntos los podemos denotar de manera explicita o de manera implicita mediante alguna propiedad, por ejemplo:

1. Anotar directamente los elementos. $A = \lbrace -2,-1\rbrace$
2. Describir el conjunto con palabras. $A$ es el conjunto de las soluciones de la ecuación $x^2+3x+2=0$
3. Escriir las carácteristicas del conjunto. $A = \lbrace x\in\mathbb{R}\mid x^2+3x+2=0\rbrace=\lbrace -1,-2\rbrace$

Los conjuntos pueden ser de diferentes naturaleza, pues si recordamos la definición solo necesitamos una colección de objetos, por ejemplo, si se tienen un grupos de estudiantes con 15 mujeres y 10 hombres podemos asociar un conjunto a este grupo

$G = \lbrace m_1,m_2\ldots, m_{15}, h_1, h_2,\ldots, h_{10} \rbrace$

### Diagramas de Venn

Los diagramas de Venn ([Jhon Venn](http://www-history.mcs.st-and.ac.uk/Biographies/Venn.html) siglo XIX) son un método gráfico para represenrar proposiciones categoricas y probar su validez (sea lo que signifique eso), pensemos en los digramas de ven como figuras geometricas que representan conjuntos y las **relaciones** entre estos. 

### Operaciones con conjuntos

<div class='def'>
<strong>Subconjunto</strong> 
<p>
Sean $A$, $B$ conjuntos. Decimos que $A$ es un subconjunto de $B$ y escribimos $A\subseteq B$, si $\omega\in A$ $\implies$ $\omega\in B$ (lo cual leemos $\omega\in A$ implica $s\in B$).
</p>
</div>

También lo podemos espresar como $B \supseteq A$. 
 
 <div class='def'>
<strong>Igualdad de conjuntos</strong> 
<p>
Sean $A$, $B$ conjuntos. Decimos que $A=B$ $\iff $  $A\subseteq B$ y $B\subseteq A$. 
</p>
</div>
  


 <div class='def'>
<strong>Subconjunto propio</strong> 
<p>

Sean $A$, $B$ conjuntos. Decimos que $A$ es un subconjunto propio de $B$ si  $A\subseteq B$ y $A\neq B$, este caso lo denotamos como $A\subset B$
</p>
</div>
 



 <div class='def'>
<strong>Conjunto vacio</strong> 
<p>
El conjunto vacio no tienen elementos, se denota por $\varnothing$ y cumple $\varnothing \subseteq A$ para todo conjunto $A$.
</p>
</div>
 


 <div class='def'>
<strong>Unión</strong> 
<p>
Sean $A$, $B$ subconjuntos de un conjunto $\Omega$. Definimos la **unión** de $A$ y $B$ por 
$$A \cup B=\lbrace \omega\in \Omega\mid \omega\in A,\mbox{ o } \omega\in B \rbrace$$
</p>
</div>
 


 <div class='def'>
<strong>Intersección</strong> 
<p>
Sean $A$, $B$ subconjuntos de un conjunto $\Omega$. Definimos la intersección de $A$ y $B$ por 
$$A \cap B=\lbrace \omega\in \Omega\mid \omega\in A,\mbox{ y } \omega\in B \rbrace$$
</p>
</div>
 


 <div class='def'>
<strong>Diferencia</strong> 
<p>
Sean $A$, $B$ subconjuntos de un conjunto $\Omega$. Definimos la diferencia de $A$ menos $B$ por 
$$A-B=\lbrace \omega\in \Omega\mid \omega\in A,\mbox{ y } \omega\notin B \rbrace= A \cap B^c$$
</p>
</div>
 


 <div class='def'>
<strong>Diferencia Simetrica</strong> 
<p>
Sean $A$, $B$ subconjuntos de un conjunto $\Omega$. Definimos la diferencia simetrica de $A$ con $B$ por 
$$A\Delta B=\lbrace \omega\in \Omega\mid \omega\in A \mbox{ y } \omega\notin B, \mbox{ o },\omega\in B \mbox{ y } \omega\notin A   \rbrace= (A-B) \cup (B-A)$$
</p>
</div>
 


 <div class='def'>
<strong>Complemento</strong> 
<p>
Sea $A$ un subconjunto de un conjunto $\Omega$. Definimos el **complemento** de $A$ por 
$$A^c=\lbrace \omega\in \Omega\mid \omega\notin A \rbrace$$
</p>
</div>
 



 <div class='def'>
<strong>Producto cartesiano</strong> 
<p>
Sean $A$, $B$ subconjuntos de un conjunto $\Omega$. Definimos el **producto cartesiano** de $A$ cruz $B$ por 
$$A\times B=\lbrace (a,b)\mid a\in A,\mbox{ y } b\notin B \rbrace$$
</p>
</div>
 


El producto cartesiano no conmuta, pues $(a,b)\neq (b,a)$, presisamente el uso de $()$ quiere denotar este echo. Por ejmplo si $A = \lbrace 0,2\rbrace$ y  $B = \lbrace 1,4\rbrace$, entonces
$$A\times B = \lbrace (0,1), (0,4), (2,1), (2,4) \rbrace$$

el cual es distinto a 
$$B\times A = \lbrace (1,0), (4,0), (1,2), (4,2) \rbrace$$

Un producto cartesiano con el que estamos familiarizados es el de $$\mathbb{R}\times\mathbb{R}=\mathbb{R}^2=\lbrace (x,y) \mid x\in\mathbb{R}, y\in\mathbb{R}\rbrace$$ que incluso tiene una interpretación geometrica, el plano cartesiano. 

Si $|A|=n$ y $|B|=m$, entonces $|A\times B|=nm$.

Si tenemos $n$ conjuntos, $A_1,\ldots, A_n$, se de fine el porducto cartesiano como 

$$A_1\times\ldots\times A_n = \lbrace (a_1,\ldots,a_n) \mid a_i\in A_i,i=1,\ldots, n\rbrace$$

 <div class='def'>
<strong>Conjunto Potencia</strong> 
<p>
Sea $A$ un conjunto. Definimos el **conjunto potencia** denotado por $\mathcal{P}(A)$, como el conjunto de subconjuntos de $A$
</p>
</div>
 


Sean $A$, $B$, $C$ subcojuntos de un conjunto $\Omega$. Tenemos:

**Propiedades de la Unión**

- $A\cup \varnothing = A$

- $A\cup A = A$

- $A\subseteq A\cup B$

- $A\cup \Omega = \Omega$

- $A\subseteq C$, $B\subseteq C$ $\implies$ $A\cup B\subseteq C$

- $A \cup A^c = \Omega$

- $A \cup B = B \cup A$ (_comutatividad_)

- $\left(A \cup B\right) \cup C = A \cup \left(B \cup C\right)$ (_asociatividad_)

- $A \cup \left(B \cap C\right) = \left(A \cup B\right) \cap \left(A \cup C\right)$ (_distributividad_)

**Propiedades de la intersección**

- $A \cap \Omega = A$

- $A \cap \varnothing =  \varnothing$

- $A \cap A = A$

- $A \cap A^c = \varnothing$

- $A \cap B = B \cap A$ (_comutatividad_)

- $\left(A \cap B\right) \cap C = A \cap \left(B \cap C\right)$ (_asociatividad_)

- $A \cap \left(B \cup C\right) = \left(A \cap B\right) \cup \left(A \cap C\right)$ (_distributividad_)
 

**Leyes de De Morgan**

(a) $\left(A \cap B\right)^c = A^c \cup B^c$

(b) $\left(A \cup B\right)^c = A^c \cap B^c$

**Propiedades del complemento**

- $\varnothing^c = \Omega$

- $\Omega^c = \varnothing$

- $(A^c)^c = A$
